# Redimensionamento de Imagem

```python
from PIL import Image  # Apenas para abrir e salvar a imagem, não para processamento

def converter_cinza_e_binaria(caminho_imagem, limiar=128):
  #Abrir imagem
  imagem = Image.open(caminho_imagem)
  largura, altura = imagem.size
  #Criando listas para as novas imagens
  imagem_cinza = Image.new('L',(largura, altura)) #L é modo de cinza
  imagem_binaria = Image.new('1',(largura, altura)) # 1 é modo binário

  for x in range(largura):
    for y in range(altura):
      r, g, b = imagem.getpixel((x,y))[:3]
      #Convertendo imagem para tons e cinza
      cinza = int(0.299*r + 0.587*g + 0.114*b)
      imagem_cinza.putpixel((x,y),cinza)
      #Convertendo imagem para binária
      if cinza < limiar:
        imagem_binaria.putpixel((x,y),0)
      else:
        imagem_binaria.putpixel((x,y),1)
  #Salvando as imagens
  imagem_cinza.save("/content/imagem_cinza.png")
  imagem_binaria.save("/content/imagem_binaria.png")
  print("Imagens salvas em /content/imagem_cinza.png e /content/imagem_binaria.png")

#Utilizando a função
converter_cinza_e_binaria("/content/Rodotrem.jpg",limiar=128)
```
